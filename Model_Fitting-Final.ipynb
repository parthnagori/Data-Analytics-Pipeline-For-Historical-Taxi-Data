```python
cols = ['pickup_zipcode', 'dropoff_zipcode',
         'pickup_day', 'pickup_hour', 'pickup_day_of_week', 'pickup_month',
         'pickup_year', 'passenger_count', 'trip_time', 'trip_distance', 'fare_amount']

features_fare = ['pickup_zipcode', 'dropoff_zipcode',
       'pickup_day', 'pickup_hour', 'pickup_day_of_week', 'pickup_month',
       'pickup_year', 'passenger_count', 'trip_distance', 'trip_time']

features_time = ['pickup_zipcode', 'dropoff_zipcode',
       'pickup_day', 'pickup_hour', 'pickup_day_of_week', 'pickup_month',
       'pickup_year', 'passenger_count', 'trip_distance']
```

```python
def standardize_data(df, features):
    sc = StandardScaler()
    temp = sc.fit_transform(df[features])
    df[features] = temp
    return df, sc
```

```python
def get_final_dataframe(df, cols):
    df = pd.read_csv("preprocessed.csv", sep=",")
    df = df[cols]
    return df    
```

```python
def label_encode_columns(df):
    le_pick_zip = LabelEncoder()
    le_pick_zip.fit(zip_code_list)
    df['pickup_zipcode'] = le_pick_zip.transform(df['pickup_zipcode']) 

    le_drop_zip = LabelEncoder()
    le_drop_zip.fit(zip_code_list)
    df['dropoff_zipcode'] = le_drop_zip.transform(df['dropoff_zipcode'])

    le_year = LabelEncoder()
    le_year.fit([2013, 2014, 2015, 2016, 2017, 2018])
    df['pickup_year'] = le_year.transform(df['pickup_year'])
    
    return df, le_pick_zip, le_drop_zip, le_year
```

```python
def train_model(df, model_type):
    if not os.path.exists("models"):
        os.makedirs("models")
    df = get_final_dataframe(df, cols)
    df, le_pick_zip, le_drop_zip, le_year = label_encode_columns(df)
    pickle.dump(le_pick_zip, open(le_paths['le_pzip'], 'wb+'))
    pickle.dump(le_drop_zip, open(le_paths['le_dzip'], 'wb+'))
    pickle.dump(le_year, open(le_paths['le_py'], 'wb+'))
    if model_type == "fare":        
        scaled_df, sc = standardize_data(df.copy(), features_fare)
        train_x, test_x, train_y, test_y = train_test_split(scaled_df[features_fare], scaled_df['fare_amount'], test_size=0, shuffle=True)
        model = RandomForestRegressor()
        model.fit(train_x, train_y)
        pickle.dump(model, open(fare_paths['model'], 'wb+'))
        pickle.dump(sc, open(fare_paths['sc_object'], 'wb+'))
        
    elif model_type == "time":
        scaled_df, sc = standardize_data(df.copy(), features_time)
        train_x, test_x, train_y, test_y = train_test_split(scaled_df[features_time], scaled_df['trip_time'], test_size=0, shuffle=True)
        model = RandomForestRegressor()
        model.fit(train_x, train_y)
        pickle.dump(model, open(time_paths['model'], 'wb+'))
        pickle.dump(sc, open(time_paths['sc_object'], 'wb+'))
    
```

```python
def perform_cv(train_x, train_y):
    seed = 7
    # prepare models
    models = []

    regressors = [
        ('KNN', KNeighborsRegressor(3)),
        ('LR', LinearRegression()),
        ('DT', DecisionTreeRegressor(max_depth=5)),
        ('RF', RandomForestRegressor(max_depth=5, n_estimators=200, max_features='auto'))
        ]
    # evaluate each model in turn
    results = []
    names = []
    #     scoring = 'neg_mean_absolute_error'
    scoring = 'neg_mean_squared_error'
    for name, model in regressors:
        print(name)
        kfold = model_selection.KFold(n_splits=5, random_state=seed)
        cv_results = model_selection.cross_val_score(model, train_x, train_x, cv=kfold, scoring=scoring, verbose=2)
        results.append(cv_results)
        names.append(name)
        msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
        print(msg)
    # boxplot algorithm comparison
    fig = plt.figure()
    fig.suptitle('Algorithm Comparison')
    ax = fig.add_subplot(111)
    plt.boxplot(results)
    ax.set_xticklabels(names)
    plt.show()
```
